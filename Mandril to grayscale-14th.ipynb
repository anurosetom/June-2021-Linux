```python
from skimage.feature import corner_harris,corner_peaks
import skimage.io as io
from skimage.color import rgb2gray
from skimage.exposure import equalize_hist
import matplotlib.pyplot as plt
```

```python
mandril=io.imread('/home/anu/Downloads/mandrill.png')
mandril[0].shape
#3=rgb
plt.imshow(mandril)
```

```python
#color image to gray scale to reduce the matrix columns
#color images have -3 matrix (for r,g & b)
#For gray scale->only 1 matrix,so we convert color image to grayscale
mandril=equalize_hist(rgb2gray(mandril))
mandril
plt.gray()
plt.imshow(mandril)
```

```python
corners=corner_peaks(corner_harris(mandril),min_distance=2)
corners.shape
```

```python
#corners
```

```python
import numpy as np

def show_corners(corners,image):#function defined for plotting corners
    fig1=plt.figure()
    plt.gray()
    
    plt.imshow(image)
    y_corner,x_corner=zip(*corners)#to make corners as iterable we use zip
    #to group data point
    
    plt.plot(x_corner,y_corner,'or')#or->orange
    plt.xlim(0,image.shape[1])#image size is actually 480 X 480
    #so depend on image size, we takes x,y values to plot on image
    plt.ylim(image.shape[0],0)
    fig1.set_size_inches(np.array(fig1.get_size_inches())*2)#to increase size not necessary

show_corners(corners,mandril)
```
