```python
import cv2
cv2.__version__
```

```python
#instead of using corner harris & corner peaks:
#     here we use SIFT to highlight the features in image
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

img1 = cv2.imread("/home/anu/Downloads/mandrill.png")
gray1 =cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.gray()
plt.imshow(gray1)
```

```python
#sift= cv2.xfeatures2d.SIFT_create()
sift= cv2.SIFT_create()

keypoints_1,descriptors_1=sift.detectAndCompute(img1,None)

img_1= cv2.drawKeypoints(gray1,keypoints_1,img1)
plt.imshow(img_1)
```

```python
cv2.imwrite("sift1.jpg",img_1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
