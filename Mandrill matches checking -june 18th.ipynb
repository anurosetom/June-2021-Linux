```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img1 = cv2.imread("/home/anu/Downloads/mandrill.png")
img2 =cv2.imread("/home/anu/Downloads/m1.jpeg")

img1 =cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2 =cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

figure,ax=plt.subplots(1,2,figsize=(16,8))

ax[0].imshow(img1,cmap='gray')
ax[1].imshow(img2,cmap='gray')

```

```python
sift=cv2.SIFT.create()

keypoints_1,descriptors_1=sift.detectAndCompute(img1,None)
keypoints_2,descriptors_2=sift.detectAndCompute(img2,None)

len(keypoints_1),len(keypoints_2)
```

```python
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)#Brutforce matcher algorithm used here

matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)#calculating the distance b/w objects in both images to compare the distance

img3=cv2.drawMatches(img1,keypoints_1,img2,keypoints_2,matches[:50],img2,flags=2)
#setting flags=2,only connect common features in both images
#if not settig flag=2, it will highlight so many features in both images

plt.imshow(img3)
plt.show()
```

```python
#saving image into device 
#cv2.imwrite("eiffel_swift_withouflag.jpg",img3)
cv2.imwrite("mandril_swift.jpg",img3)

cv2.waitKey(0)
cv2.destroyAllWindows
```
