```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/home/anu/Downloads/homeprice_nlp.csv',sep='\t')
df
```

# Giving integer values to place name

# Nominal variables=>categories that do not have numerical order between them
#here place names are in nominal category(because we can't rank them based on place name

# Ordinal Variables=>Exactly having numerical ordering between them
#egs:low,medium,high=>these can be arranged in an order based on its value

```python
#One hot encoding is done using dummy variable
dummies=pd.get_dummies(df.town)#getting dummy variable for column "town"
dummies
```

# This method is only for small dataset.For ordinal values,1 hot encoding is suitable(have rank based system)

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town'],axis='columns')
final
```

# There is collinearity b/w last 2 columns, so we can drop one of them
# 1,0=>0 for west_windsor
# 0,0=>1

```python
final=final.drop(['west windsor'],axis=1)
final
```

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[3400,0,0]])#0,0 denotes west windsor,here we are predicting west windsor's price
```

```python
model.predict([[3400,0,1]])#robinsville prediction =0,1
```
